```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\Elie\Desktop\Go\Auction.csv')
```

```python
df.set_index('Mois')
```

```python
Vol_2019 = df[df.columns[1]][2:].to_list()
Vol_2020 = df[df.columns[3]].to_list()
Vol_2021 = df[df.columns[5]].to_list()
Vol_2022 = df[df.columns[7]].to_list()
Vol_2023 = df[df.columns[9]][:5].to_list()

prix_2019 = df[df.columns[2]][2:].to_list()
prix_2020 = df[df.columns[4]].to_list()
prix_2021 = df[df.columns[6]].to_list()
prix_2022 = df[df.columns[8]].to_list()
prix_2023 = df[df.columns[10]][:5].to_list()
```

```python
vol = Vol_2019 + Vol_2020 + Vol_2021 + Vol_2022 + Vol_2023 
prix = prix_2019 + prix_2020 + prix_2021 + prix_2022 + prix_2023
```

```python
def show_plot(
    vol,
    figsize=(15, 10),
    color="blue",
    linestyle="-",
    xlabel="Mois",
    ylabel="Volume total moyen vendu",
    label=None,
):
    plt.figure(figsize=figsize)
    plt.plot(vol, color=color, label=label)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend(loc=2)
```

**Années d'études : Mars 2019 - Mai 2023**

```python
show_plot(vol,label = 'All data')
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
decompose = seasonal_decompose(vol,model = 'additive',period = 12)
fig = decompose.plot()
fig.set_size_inches(12,6)
```

```python
plt.figure(figsize=(18, 10)) 
plt.plot(decompose.seasonal)
plt.xlabel('Mars 2019 - Mai 2023',fontsize=14)
plt.ylabel('Volume total moyen vendu',fontsize=14)
plt.title('Détection de la saisonnalité des volumes',fontsize=14)
```
