```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\Elie\Desktop\Go\Auction.csv')
```

```python
df.set_index('Mois')
```

```python
Vol_2019 = df['Vol_2019'][2:].to_list()
Vol_2020 = df['Vol_2020'].to_list()
Vol_2021 = df['Vol_2021'].to_list()
Vol_2022 = df['Vol_2022'].to_list()
Vol_2023 = df['Vol_2023'][:5].to_list()

Price_2019 = df['Price_2019'][2:].to_list()
Price_2020 = df['Price_2020'].to_list()
Price_2021 = df['Price_2021'].to_list()
Price_2022 = df['Price_2022'].to_list()
Price_2023 = df['Price_2023'][:5].to_list()
```

```python
X = Vol_2020 + Vol_2021 + Vol_2022 + Vol_2023 
y = Price_2020 + Price_2021 + Price_2022 + Price_2023
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
decompose = seasonal_decompose(X,model = 'additive',period=12)
fig = decompose.plot()
fig.set_size_inches(12,6)
```

```python
fig_2 = pd.Series(decompose.seasonal).plot(figsize=(18, 10))
plt.title('Seasonal decomposition',fontsize=16)
plt.ylabel('Volume', fontsize=16)
plt.xlabel('Dates', fontsize=16)
```
