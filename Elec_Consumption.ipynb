```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\Elie\Desktop\Go\Elec.csv')
```

```python
df
```

```python
date = [i for i in range(1,13)]
date_11 = [i for i in range(1,12)]
date_10 = [i for i in range(1,11)]
date_5 = [i for i in range(1,6)]
```

```python
Conso_2019 = df['Conso_2019'][2:].to_list()
Conso_2020 = df['Conso_2020'].to_list()
Conso_2021 = df['Conso_2021'][:-1].to_list()

Price_2019 = df['Price_2019'][2:].to_list()
Price_2020 = df['Price_2020'].to_list()
Price_2021 = df['Price_2021'][:-1].to_list()
```

```python
fig, ax = plt.subplots(3,1, figsize=(10,8))
ax2 = np.array([a.twinx() for a in ax.ravel()]).reshape(ax.shape)

ax[0].plot(date_10,Price_2019)
ax2[0].plot(date_10,Conso_2019, color="red")

ax[0].set_ylabel("Price EUR/Mwh", color = "blue")
ax2[0].set_ylabel("Total Consumption", color = 'red')

ax[0].set_title("2019")

ax[1].plot(date,Price_2020)
ax2[1].plot(date,Conso_2020, color="red")

ax[1].set_ylabel("Price EUR/Mwh", color = "blue")
ax2[1].set_ylabel("Total Consumption", color = 'red')

ax[1].set_title("2020")
plt.tight_layout()

ax[2].plot(date_11,Price_2021)
ax2[2].plot(date_11,Conso_2021, color="red")

ax[2].set_ylabel("Price EUR/Mwh", color = "blue")
ax2[2].set_ylabel("Total Consumption", color = 'red')

ax[2].set_title("2021")
plt.tight_layout()
```
